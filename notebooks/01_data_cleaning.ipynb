Missing Value Analysis

```python
# ============================================
# Missing Value Analysis
# ============================================

missing = (
    df.isnull()
      .sum()
      .to_frame(name="Missing Values")
)

missing["Missing %"] = (
    missing["Missing Values"] / len(df) * 100
).round(2)

missing = (
    missing
    .sort_values("Missing %", ascending=False)
)

missing.head(60)
```

```python
# ============================================
# Final Business Columns
# ============================================

selected_columns = [

    # Loan Information
    "loan_amnt",
    "term",
    "int_rate",
    "installment",
    "grade",
    "sub_grade",
    "purpose",
    "loan_status",
    "issue_d",

    # Customer Information
    "annual_inc",
    "emp_length",
    "home_ownership",
    "verification_status",

    # Credit Profile
    "fico_range_low",
    "fico_range_high",
    "dti",
    "open_acc",
    "pub_rec",
    "revol_bal",
    "revol_util",
    "total_acc",
    "delinq_2yrs",
    "inq_last_6mths",

    # Additional Financial Features
    "earliest_cr_line",
    "mort_acc",
    "tot_cur_bal",
    "total_bc_limit",
    "tot_hi_cred_lim",

    # Location
    "addr_state",

    # Application
    "application_type"
]

len(selected_columns)
```

```python
# ============================================
# Create Selected Dataset
# ============================================

df_selected = df[selected_columns].copy()

print(df_selected.shape)

df_selected.head()
```

```python
# ============================================
# Save Selected Dataset
# ============================================

output_path = "../data/processed/lending_club_selected_columns.csv"

df_selected.to_csv(
    output_path,
    index=False
)

print("Dataset saved successfully!")
```

```python
print(
    f"Memory Usage: "
    f"{df_selected.memory_usage(deep=True).sum()/1024**2:.2f} MB"
)
```

```python
df_selected.info()
```

```python
df_selected.dtypes
```

```python
df_selected["term"] = (
    df_selected["term"]
    .str.replace(" months", "", regex=False)
)

df_selected["term"] = pd.to_numeric(
    df_selected["term"],
    errors="coerce"
)
```

```python
percentage_columns = [
    "int_rate",
    "revol_util"
]

for col in percentage_columns:
    df_selected[col] = (
        df_selected[col]
        .astype(str)
        .str.replace("%", "", regex=False)
    )

    df_selected[col] = pd.to_numeric(
        df_selected[col],
        errors="coerce"
    )
```

```python
date_columns = [
    "issue_d",
    "earliest_cr_line"
]

for col in date_columns:
    df_selected[col] = pd.to_datetime(
        df_selected[col],
        format="%b-%Y",
        errors="coerce"
    )
```

```python
# ============================================
# Missing Values in Selected Dataset
# ============================================

missing = (
    df_selected.isnull()
    .sum()
    .to_frame("Missing Values")
)

missing["Missing %"] = (
    missing["Missing Values"] / len(df_selected) * 100
).round(2)

missing = missing[missing["Missing Values"] > 0]

missing.sort_values("Missing %", ascending=False)
```

```python
# ============================================
# Handle Missing Values
# ============================================

# Fill categorical column
df_selected["emp_length"] = df_selected["emp_length"].fillna("Unknown")

# Fill numerical columns with median
median_columns = [
    "tot_hi_cred_lim",
    "tot_cur_bal",
    "total_bc_limit",
    "mort_acc",
    "dti",
    "revol_util"
]

for col in median_columns:
    df_selected[col] = df_selected[col].fillna(
        df_selected[col].median()
    )

# Remove remaining rows with missing values
df_selected.dropna(inplace=True)

print(df_selected.shape)
```

# Feature Engineering.

## Feature 1: Credit Score

```python
# ============================================
# Feature 1: Credit Score
# ============================================

df_selected["credit_score"] = (
    (
        df_selected["fico_range_low"] +
        df_selected["fico_range_high"]
    ) / 2
)

df_selected["credit_score"].describe()
```

## Feature 2: Income Group

```python
# ============================================
# Feature 2: Income Group
# ============================================

income_bins = [
    0,
    50000,
    100000,
    200000,
    float("inf")
]

income_labels = [
    "Low",
    "Middle",
    "High",
    "Very High"
]

df_selected["income_group"] = pd.cut(
    df_selected["annual_inc"],
    bins=income_bins,
    labels=income_labels
)

df_selected["income_group"].value_counts()
```

## Feature 3: Loan Amount Group

```python
loan_bins = [
    0,
    5000,
    15000,
    30000,
    float("inf")
]

loan_labels = [
    "Small",
    "Medium",
    "Large",
    "Very Large"
]

df_selected["loan_amount_group"] = pd.cut(
    df_selected["loan_amnt"],
    bins=loan_bins,
    labels=loan_labels
)

df_selected["loan_amount_group"].value_counts()
```

## Feature 4: Interest Rate Group

```python
rate_bins = [
    0,
    8,
    13,
    18,
    float("inf")
]

rate_labels = [
    "Low",
    "Moderate",
    "High",
    "Very High"
]

df_selected["interest_rate_group"] = pd.cut(
    df_selected["int_rate"],
    bins=rate_bins,
    labels=rate_labels
)

df_selected["interest_rate_group"].value_counts()
```

## Feature 5: Issue Year

```python
df_selected["issue_year"] = (
    df_selected["issue_d"]
    .dt.year
)

df_selected["issue_year"].value_counts().sort_index()
```

```python
new_features = [
    "credit_score",
    "income_group",
    "loan_amount_group",
    "interest_rate_group",
    "issue_year"
]

df_selected[new_features].head()
```

```python
df_selected.to_csv(
    "../data/processed/lending_club_clean.csv",
    index=False
)

print("✅ Feature engineered dataset saved successfully!")
```
